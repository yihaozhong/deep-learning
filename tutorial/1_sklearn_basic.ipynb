###**What is sklearn and how does it support Machine Learning?**

---

```sklearn``` or ```scikit-learn``` is a free and open source machine learning library built to have most of the features/functions used in ML. It is one of the most famous library. Around one-quarter of the people who do Machine Learning use ```scikit-learn``` on a regular basis (as per one of the core ```sklearn``` contributors).

It presents easy-to-use functionalities consisting standard machine learning algorithms used in predictable modeling. It incorporates implementations of algorithms like classification, regression and clustering algorithms including support-vector machines, random forests, gradient boosting, k-means. It is mainly written in Python and integrates well with other famous libraries like [```Pandas```](https://pandas.pydata.org/), [```NumPy```](https://numpy.org/), [```Matplotlib```](https://matplotlib.org/).

### **Few terms commonly used in ML: train, test, features, target, sample.**

---
- **Train**: This is subset of the dataset which is used for ML model training purposes. The Model tries to optimize and learn based on this subset of the data.
- **Test**: This is a subset of the dataset that is used for testing the trained model. Important note: Test ∩ Train = Φ i.e. there should be no data point that is present in both test and train set.
- **Features**: A feature is a quantifiable characterstic of the dataset. Ideal features for an ML task should be differentiating, and informative. These are also called independent variables.
- **Target**: Target variable is the characterstic/feature of the dataset that we try to predict. The model tries to predict an estimate of the target variable. It is called dependent variable.
- **Sample**: A sample is a subset of the train or test dataset.
- **Loss**: It is a numerical value associated with each estimate which is a penalty for a incorrect estimate. Some of the loss functions are [Mean Squared Error](https://en.wikipedia.org/wiki/Mean_squared_error), [Mean Absolute Error](https://en.wikipedia.org/wiki/Mean_absolute_error), [Cross Entropy Loss](https://en.wikipedia.org/wiki/Cross_entropy).

### **Load Data**
---

The sklearn.datasets package features helpers to fetch larger datasets commonly used by the machine learning community to benchmark algorithms on data that comes from the ‘real world’.

There are three main kinds of dataset interfaces that can be used to get datasets depending on the desired type of dataset.

- The dataset loaders: They can be used to load small standard datasets, described in the [Toy datasets section](https://scikit-learn.org/stable/datasets/toy_dataset.html) of the documentation.

- The dataset fetchers: They can be used to download and load larger datasets, described in the [Real world datasets section](https://scikit-learn.org/stable/datasets/real_world.html) of the documentation.

- The dataset generation functions: They can be used to generate controlled synthetic datasets. These functions return a tuple (X, y) consisting of a n_samples * n_features numpy array X and an array of length n_samples containing the targets y.

### Split into Train/Test Sets
---
The train-test split is a technique for evaluating the performance of a machine learning algorithm. The procedure involves taking a dataset and dividing it into two subsets.
It can be used for classification or regression problems and can be used for any supervised learning algorithm.

- Train Subset: used to fit the model and is referred to as the training dataset.

- Test Subset: input element of the dataset is provided to the model, then predictions are made and compared to the expected values. This second dataset is referred to as the test dataset.

The objective is to estimate the performance of the machine learning model on new data: data not used to train the model.

````python
# Store the feature data
X = dataset.data
# store the target data
y = dataset.target
# split the data using Scikit-Learn's train_test_split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, shuffle= True)

print("Total size of the sample:", len(X))
print("Size of Training set:", len(X_train))
print("Size of Test set:", len(X_test))
````

### **Walkthrough example**

---

Let us walkthrough a real life example to illustrate the working of the Linear Regression model.

We'll be using the California Housing dataset, which contains information about different houses in California. The dataset can be directly accessed from the `sklearn` library and contains 20640 samples and 8 feature variables. The objective is to predict the value of prices of the house using the given features.

````python
# Import the necessary dependencies

import sklearn.datasets as datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import sklearn.metrics as metrics
````

````python
# Load the dataset and print dataset info

house = datasets.fetch_california_housing()

print("The data shape of house is {}".format(house.data.shape))
print("The number of feature in this data set is {}\n\n".format(
    house.data.shape[1]))

print(house.DESCR)
````

````python
print(house.keys())
````

````python
# Make the train and test sets

train_x, test_x, train_y, test_y = train_test_split(house.data,
                                                    house.target,
                                                    test_size=0.2,
                                                    random_state=42)

print("The first five samples {}".format(train_x[:5]))
print("The first five targets {}".format(train_y[:5]))
print("The number of samples in train set is {}".format(train_x.shape[0]))
print("The number of samples in test set is {}".format(test_x.shape[0]))
````

````python
# Instantiate a linear regression model, and fit the model to the data

lr = LinearRegression()
lr.fit(train_x, train_y)
````

````python
# Make predictions on the test set

pred_y = lr.predict(test_x)
print("The first five prediction {}".format(pred_y[:5]))
print("The real first five labels {}".format(test_y[:5]))
````

````python
# Get the MSE of the predictions and the ground truth
mse = metrics.mean_squared_error(test_y, pred_y)
print("Mean Squared Error {}".format(mse))
````

### How to evaluate the performance of models in Sklearn
---

We use the Sklearn.metrics class to use in-built functions for most of these functions.

**Mean Squared Error:** It is defined as the average squared difference between the predicted values and the actual values.

**Mean Absolute Error:** It is an average of the all absolute errors
  

````python
from sklearn import metrics

# Mean Squared Error
mse = metrics.mean_squared_error(test_y, pred_y)
print("Mean Squared Error:", mse)

# Mean Absolute Error
mae = metrics.mean_absolute_error(test_y, pred_y)
print("Mean Absolute Error:", mae)

````

**Confusion Matrix:**
A confusion matrix is an N X N matrix, where N is the number of classes being predicted. For the problem in hand, we have N=2, and hence we get a 2 X 2 matrix. Here are a few definitions, you need to remember for a confusion matrix :

- Accuracy : the proportion of the total number of predictions that were correct.
- Positive Predictive Value or Precision : the proportion of positive cases that were correctly identified.
- Negative Predictive Value : the proportion of negative cases that were correctly identified.
- Sensitivity or Recall : the proportion of actual positive cases which are correctly identified.
- Specificity : the proportion of actual negative cases which are correctly identified.
